```python
import pandas as pd

df = pd.read_csv("../data/raw/creditcard.csv")
df.head()

```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Replace Isolation Forest output (-1, 1) to (1, 0)
df['if_anomaly'] = df['if_anomaly'].replace({-1:1, 1:0})

print(confusion_matrix(df['Class'], df['if_anomaly']))
print(classification_report(df['Class'], df['if_anomaly']))
```

```python
from sklearn.ensemble import IsolationForest

X = df.drop('Class', axis=1)

model = IsolationForest(contamination=0.01, random_state=42)

df['if_anomaly'] = model.fit_predict(X)
```

```python
df['if_anomaly'] = df['if_anomaly'].replace({-1:1, 1:0})
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(df['Class'], df['if_anomaly']))
print(classification_report(df['Class'], df['if_anomaly']))
```
