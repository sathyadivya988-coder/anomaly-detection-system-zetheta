```python
import pandas as pd

df = pd.read_csv(r"C:\Users\it272\OneDrive\Documents\anamoly-detection-system\data\raw\creditcard.csv")

df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df_scaled = df.copy()
df_scaled[['Amount', 'Time']] = scaler.fit_transform(df[['Amount', 'Time']])
```

```python
X = df_scaled.drop('Class', axis=1)
```

```python
from sklearn.ensemble import IsolationForest

model = IsolationForest(
    n_estimators=100,
    contamination=0.002,
    random_state=42
)

model.fit(X)
```

```python
df['iforest_pred'] = model.predict(X)
```

```python
df['iforest_anomaly'] = (df['iforest_pred'] == -1).astype(int)
```

```python
df['iforest_anomaly'].value_counts()
```

```python
pd.crosstab(df['Class'], df['iforest_anomaly'])
```
